```python
#exploring the data for possible regression
#first case adresses the electorate against the valid votes cast

y = Dom1['Electorate']
x1 = Dom1['Valid votes cast']
plt.scatter(x1,y)
plt.xlabel('Valid votes cast',fontsize=20)
plt.ylabel('Electorate', fontsize=20)
```

```python
#plotting the regression line
plt.scatter(x1,y)
yhat = -1.143e+04 + 1.5160*x1
fig = plt.plot(x1,yhat, lw=4, c= 'orange', label = 'regression line')
plt.xlabel('Valid votes cast',fontsize=20)
plt.ylabel('Electorate', fontsize=20)
plt.show()
```

```python
#Defining x and y for the model

x = Model1.drop(['Votes cast'],axis=1).values
print(x)

```

```python
y = Model1['Votes cast'].values
print(y)
```

```python
#splitting the dataset into training and test sets

from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test\
= train_test_split(x,y,test_size=0.3, random_state=0)

#Training the model on the training set
from sklearn.linear_model import LinearRegression 

#fitting the model
ml=LinearRegression()
ml.fit(x_train,y_train)
```

```python
#Below are the coefficients of Bn's (coefficients) given in -
#the order of the column names in the above sample table of the model
ml.coef_
```

```python
#The bnode, which gives the intercept is indicated below
ml.intercept_
```

```python
#predicting the test set results 
PredictedY=ml.predict(x_test)

#Viewing the predicted results of the 3rd record 
ml.predict([[0.109185, 7.736261, 0.071972, -1.086427]])
#compared to 161218
```

```python
#MODEL EVALUATION

#r2 below gives the rsquared of the regression model
from sklearn.metrics import r2_score
r2_score(y_test,PredictedY)
```

```python
#From the above assessment, the model is 30 percent accurate

#plotting the results
plt.scatter(y_test,PredictedY)
plt.xlabel('Actual',fontsize=20)
plt.xlabel('Predicted',fontsize=20)
plt.title('Actual Vs Predicted', fontsize=20)
sns.set(rc={"figure.figsize": (4, 3)})
```

```python
#Outputting the Predicted Values (the first 15)
Prediction_df= pd.DataFrame({'Actual Value':y_test,'Predicted Value':PredictedY,'Difference':y_test-PredictedY})
Prediction_df[0:15]
```
